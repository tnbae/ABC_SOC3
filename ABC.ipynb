```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
entry_data=pd.read_csv('~/aiffel/data_analysis_basic/data/entry_data.csv')
sales_data=pd.read_csv('~/aiffel/data_analysis_basic/data/sales_data.csv')
item_data=pd.read_csv('~/aiffel/data_analysis_basic/data/item_data.csv')
item_entry_data=pd.read_csv('~/aiffel/data_analysis_basic/data/item_entry_data.csv')
```

```python
temp_var = sales_data
```

```python
temp_var['date'] = sales_data['결제일시'].apply(lambda i: str(i[:7]))
temp_var
```

```python
temp_var['결제금액'] = temp_var['결제금액'].str.replace(',', '')
temp_var['결제금액']
```

```python
temp_var['결제금액'] = temp_var['결제금액'].astype(int)
```

```python
monthly = temp_var.groupby('date')['결제금액'].sum()
monthly = pd.DataFrame(monthly)
monthly.reset_index(level = ['date'], inplace = True)
monthly
```

```python
sns.barplot(data = monthly, x = 'date', y = '결제금액')
```

```python
temp_var = temp_var[temp_var['판매수량'] == 1]
popular = pd.DataFrame(temp_var['상품명'].value_counts())
popular.rename(columns = {'상품명': '판매량'}, inplace = True)
popular.reset_index(level=0, inplace=True)
popular.rename(columns = {'index': '상품명'}, inplace = True)
popular
```

```python
top = popular.nlargest(10, '판매량', keep = 'first')
top
```

```python
plt.rc('font', family = 'NanumGothic')
%matplotlib inline
```

```python
sns.barplot(data = top, x = '상품명', y = '판매량')
```

```python
money = temp_var.groupby('상품명')['결제금액'].sum()
money = pd.DataFrame(money)
money.reset_index(level = ['상품명'], inplace = True)
money = pd.merge(money, popular)
money['총 매출'] = money['결제금액'] * money['판매량']
money
```

```python
most = money.nlargest(10, '총 매출', keep = 'first')
most
```

```python
sns.barplot(data = most, x = '상품명', y = '총 매출')
```

```python
temp_var = entry_data
```

```python
temp_var['금액'] = temp_var['금액'].astype(int)
```

```python
price = temp_var.groupby('요금제명')['금액'].sum()
price = pd.DataFrame(price)
price.reset_index(level = ['요금제명'], inplace = True)
price
```

```python
ticket = price.nlargest(10, '금액', keep = 'first')
ticket
```

```python
ax = sns.barplot(data = ticket, x = '요금제명', y = '금액')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
```
